## Analyzing Nulls

#### The data set has null values. This section will analyze if there is a significant difference in the null values to determine how we should treat the nulls moving forward. I will be looking at when nulls are disregarded and when nulls are valued zero. Zero is selected so that the locations/months with null will be penalized for not having data. This way we can see if null values impact our data significantly.

This was completed in the 2016-2020 dataset and the results were very similar.

## Comparing Monthly Data - Visualization

### Null Values Set to Zero, Mean Dataframe

*Filling in NaN (null values) with zero, to compute new mean:*

```python
null_zero_df = complete_df.fillna(0)
```

```python
null_zero_monthly_mean_df = null_zero_df[MONTH_COLUMNS].mean(axis=0, numeric_only=True).to_frame().reset_index().rename(columns={'index':'Month',0:'Mean'})
```

#### Standard Deviation for Null Values Set to Zero

*Calculating standard deviation and standard error for `null_zero_df` for each month (using original dataframe)*

```python
null_zero_stde_df = pd.DataFrame({
                        'STD' : null_zero_df[MONTH_COLUMNS].std(),
                        'SE' :  null_zero_df[MONTH_COLUMNS].sem(),
                    }).reset_index().rename(columns={'index':'Month'})
```

### Null Values Disregarded, Mean Dataframe

*Creating dataframe for `monthly_mean_df`*

```python
monthly_mean_df = complete_df[MONTH_COLUMNS].mean(axis=0, numeric_only=True).to_frame().reset_index().rename(columns={'index':'Month',0:'Mean'})
```

#### Standard Deviation for Null Values Disregarded

*Calculating standard deviation <span style='color:gray; font-size:.8em;'>(`.std()` automatically calculates using n-1)</span> and standard error for `monthly_mean_df` for each month (using original dataframe)*

```python
all_monthly_std_df = pd.DataFrame({
                        'STD' : complete_df[MONTH_COLUMNS].std(),
                        'SE' :  complete_df[MONTH_COLUMNS].sem(),
                    }).reset_index().rename(columns={'index':'Month'})
```

### DataFrame with Both Zero and Disregarded Null

```python
null_comparison_means_df = pd.DataFrame({
                                'Month': null_zero_monthly_mean_df['Month'],
                                'Null_Zero' : null_zero_monthly_mean_df['Mean'],
                                'Null_Disregarded' : monthly_mean_df['Mean'],
                                'Null_Zero_SE' : null_zero_stde_df['SE'],
                                'Null_Disregarded_SE' : all_monthly_std_df['SE'],
                            })
```

*Double bar graph for `comparing_monthly_df` to show trends from both mean calculations: `Null_Disregarded` and `Null_Zero`*

```python
spaced_range = np.arange(12)
width = 0.4

all_location_error = 2*all_monthly_std_df['SE']
null_zero_error = 2*null_zero_stde_df['SE']

fig = plt.figure(figsize=(22,10))
plt.bar(spaced_range, null_comparison_means_df['Null_Disregarded'], width = width,  yerr= all_location_error, label='Null_Disregarded', color='violet')
plt.bar(spaced_range + width, null_comparison_means_df['Null_Zero'], width = width, yerr= null_zero_error, label ='Null_Zero', color='royalblue')

plt.xticks(spaced_range + width/2, null_comparison_means_df['Month'], fontsize=15)
plt.title('Comparing Mean Red Light Violations For Ottawa Locations \n2016-2021',fontsize=35)
plt.xlabel('Month', fontsize=30)
plt.ylabel('Mean Red Light Violations', fontsize=30)
plt.legend(fontsize=25)
# plt.savefig('/Users/anjawu/Code/anjawu.github.io/images/pages/projects/redlight/comparing_null_means_monthly.png', bbox_inches='tight', facecolor = 'white')
plt.show()
```

This was completed in the 2016-2020 dataset and the results were very similar.

<span style="color:lightseagreen; font-size:1.3em;">
	As can be seen the same general trends hold true for both datasets. Moving forward, I will be working with the null disregarded dataset - looking at only the active locations because it's reasonable to assume that null means the camera wasn't working which is different than the camera working and observing 0 red lights and I wanted the data to encode this.

## Grouping by Intersection Analysis

*Sorting the complete dataframe by the intersection, along with the year that the camera was installed to account for the multiple cameras that can be at an intersections. Every other column will have a mean actioned on the matching rows.*

```python
group_by_intersection_year = complete_df.groupby(['INTERSECTION','CAMERA_INSTALL_YEAR'], axis=0).mean().reset_index()
# group_by_intersection_year
```

## Calculations of money loss for bottom locations

*Create function that will calculate the negative profit for the locations that do not breakeven*

We want to create a table that will have the profit displayed for each location that does not breakeven and show the profit for each year it was operational.

- [x] identify the locations that have had a year that have not broken even (yearly: 204, monthly: 17 violations)
- [x] pull monthly mean
- [x] calculate profit (multiple by revenue of ticket 205 minus the operations cost 53000 annually or 4417 monthly)
- [x] put each location with each year profit in a dictionary to put in a df to look like this:

| Year      | 2016 | 2017 | ... | 2021 |
| ----------- | ----------- |  ----------- | ----------- |  ----------- | 
| Location 1  | Monthly profit | Monthly profit | ... |  Monthly profit |
| Location 2 | Monthly profit | Monthly profit | ... |  Monthly profit |
| ... | ... | ... | ... | ... |

- [x] highlight the profits that fall below 0 in red
- [ ] figure out how to add install year as column maybe (look at location from initial dataframe and pull out year and put as column) 
- [x] repeat process with total violations

```python
# our dataframe has the average monthly violations for each year from 2016 to 2021

sort_by_intersection_mean_table_replacezeros = sort_by_intersection_mean_table.replace(0.0,np.NaN)

# make list of locations that have at some point between 2016-2021 not broken even for monthly mean violations
below_breakeven_locations_list = sort_by_intersection_mean_table_replacezeros[(sort_by_intersection_mean_table_replacezeros['2016']<17)
                               |(sort_by_intersection_mean_table_replacezeros['2017']<17)
                               |(sort_by_intersection_mean_table_replacezeros['2018']<17)
                               |(sort_by_intersection_mean_table_replacezeros['2019']<17)
                               |(sort_by_intersection_mean_table_replacezeros['2020']<17)
                               |(sort_by_intersection_mean_table_replacezeros['2021']<17)
                               ].index.tolist()

# list of years
years = sort_by_intersection_mean_table.columns.tolist()

column_names = ['location']+years

# create df to store profit for the monthly mean for each location each year
month_profit_df = pd.DataFrame(columns= years)

location_list = []

# cycle through each location in list
for i, location in enumerate(below_breakeven_locations_list):
    # list to store the years monthly mean profit
    yearly_month_profit = []
    # Add location 
    location_list.append(location)
    # cycle through each year
    for year in years:
        # find specific value use .at https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.at.html
        monthly_violation = sort_by_intersection_mean_table.at[location, year]
        # each violation earns $260 that goes to the city
        monthly_revenue = monthly_violation*260
        
        # calculate monthly profit (monthly earned - expenses $4,416.67)
        monthly_profit = monthly_revenue - 4416.67
        # add 2016-2021 mean monthly profit
        yearly_month_profit.append(monthly_profit)
        
        # for my sake I was curious what the annual profit would be using the monthly mean
        monthly_annual_profit = monthly_profit*12

    # appending row by row for each location.
    month_profit_df.loc[i] = yearly_month_profit
    

# Add location information to index
month_profit_df.set_index([location_list], inplace=True)

# colour change to red if profit less than 0: https://stackoverflow.com/questions/28075699/coloring-cells-in-pandas
def color_red(val):
    color = 'red' if val < 0 else 'black'
    return 'color: %s' % color


month_profit_coloured_df=month_profit_df.style.applymap(color_red)

dfi.export(month_profit_coloured_df, 'month_profit_coloured_df.png')
month_profit_coloured_df
```

```python
# our dataframe has the average monthly violations for each year from 2016 to 2021
sort_by_intersection_total_table_replacezeros = sort_by_intersection_total_table.replace(0.0,np.NaN)

# make list of locations that have at some point between 2016-2021 not broken even for monthly mean violations
below_breakeven_locations_total_list = sort_by_intersection_total_table_replacezeros[(sort_by_intersection_total_table_replacezeros['2017']<204)
                               |(sort_by_intersection_total_table_replacezeros['2018']<204)
                               |(sort_by_intersection_total_table_replacezeros['2019']<204)
                               |(sort_by_intersection_total_table_replacezeros['2020']<204)
                               |(sort_by_intersection_total_table_replacezeros['2021']<204)
                               ].index.tolist()

# list of years
years = sort_by_intersection_total_table.columns.tolist()
column_names = ['location']+years

# create df to store profit for the yearly total for each location each year
total_profit_df = pd.DataFrame(columns= years)

location_list = []

# cycle through each location in list
for i, location in enumerate(below_breakeven_locations_total_list):
    # list to store the years monthly mean profit
    yearly_total_profit = []
    # Add location 
    location_list.append(location)
    # cycle through each year
    for year in years:
        # find specific value use .at https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.at.html
        total_violation = sort_by_intersection_total_table.at[location, year]
        # each violation earns $260 that goes to the city
        total_revenue = total_violation*260
        
        # calculate monthly profit (monthly earned - expenses $53000)
        total_profit = total_revenue - 53000
        # add 2016-2021 mean monthly profit
        yearly_total_profit.append(total_profit)
       
    # appending row by row for each location.
    total_profit_df.loc[i] = yearly_total_profit
    

# Add location information to index
total_profit_df.set_index([location_list], inplace=True)

# colour change to red if profit less than 0: https://stackoverflow.com/questions/28075699/coloring-cells-in-pandas
def color_red(val):
    color = 'red' if val < 0 else 'black'
    return 'color: %s' % color

total_profit_coloured_df=total_profit_df.style.applymap(color_red)

dfi.export(total_profit_coloured_df, 'total_profit_coloured_df.png')
total_profit_coloured_df
```

Therefore a new DataFrame needs to be created that combines 2020 and 2019, including:  `'INTERSECTION'`, `'CAMERA_FACING'`, `'CAMERA_INSTALL_YEAR'` all months, total number of violations and monthly mean.*

```python
twentyone_months_df = twentyone_df[['INTERSECTION', 'CAMERA_FACING', 'CAMERA_INSTALL_YEAR', 'LATITUDE', 'LONGITUDE'] +MONTH_COLUMNS].set_index(['INTERSECTION','CAMERA_FACING'])
twentyone_months_df.index = [': '.join(i) for i in twentyone_months_df.index]
twentyone_months_df['Total_Violation 2020'] = twentyone_months_df[MONTH_COLUMNS].sum(axis=1)
twentyone_months_df['Monthly_Mean 2020'] = twentyone_months_df[MONTH_COLUMNS].mean(axis=1)

twenty_months_df = twenty_df[['INTERSECTION', 'CAMERA_FACING'] +MONTH_COLUMNS].set_index(['INTERSECTION','CAMERA_FACING'])
twenty_months_df.index = [': '.join(i) for i in twenty_months_df.index]
twenty_months_df['Total_Violation 2020'] = twenty_months_df[MONTH_COLUMNS].sum(axis=1)
twenty_months_df['Monthly_Mean 2020'] = twenty_months_df[MONTH_COLUMNS].mean(axis=1)


nineteen_months_df = nineteen_df[['INTERSECTION','CAMERA_FACING'] +MONTH_COLUMNS].set_index(['INTERSECTION', 'CAMERA_FACING'])
nineteen_months_df.index = [': '.join(i) for i in nineteen_months_df.index]
nineteen_months_df['Total_Violation 2019'] = nineteen_months_df[MONTH_COLUMNS].sum(axis=1)
nineteen_months_df['Monthly_Mean 2019'] = nineteen_months_df[MONTH_COLUMNS].mean(axis=1)
```
